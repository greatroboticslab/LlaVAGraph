```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from PIL import Image
import random

warnings.filterwarnings(action="ignore")
```

```python
NUM_OF_IMAGES = 25
BASE_PATH = "baseImage.png"
```

```python
combinedData = pd.read_parquet("combinedData.gzip")
```

```python
baseImage = Image.open(BASE_PATH)
```

```python
combinedData["File"] = combinedData["File"].str.replace("\\", "-")
```

```python
px = 1 / plt.rcParams["figure.dpi"]  # pixel in inches
```

```python
def makeChartImage(x, y, baseImage):
    fig, ax = plt.subplots(figsize=(700 * px, 300 * px))
    plt.plot(x, y, color="#b43ed1")

    plt.axis("off")  # this rows the rectangular frame
    ax.get_xaxis().set_visible(False)  # this removes the ticks and numbers for x axis
    ax.get_yaxis().set_visible(False)

    plt.savefig("temp.png", bbox_inches="tight", pad_inches=0.1)
    plt.close()
    chart = Image.open("temp.png")

    new_image = baseImage

    new_image.paste(chart, (80, 200))

    return new_image
```

```python
for file in combinedData["File"].unique():
    print(file)
    trialData = combinedData[combinedData["File"] == file]

    length = len(trialData["D-Value"])
    chartType = file.split()[0]
    delta = list(trialData["D-Value"])

    deltaDValue = [0 for i in range(length)]
    deltaDisplacement = [0 for i in range(length)]
    totalDisplacement = [0 for i in range(length)]

    for i in range(1, len(delta)):
        deltaDValue[i] = delta[i] - delta[i - 1]
    for i in range(1, len(delta)):
        deltaDisplacement[i] = (deltaDValue[i] + deltaDValue[i - 1]) * 79.2
    for i in range(1, length):
        totalDisplacement[i] = deltaDisplacement[i] + totalDisplacement[i - 1]

    trialData.loc[:, "delta-d-value"] = deltaDValue
    trialData.loc[:, "delta-displacement"] = deltaDisplacement
    trialData.loc[:, "total-displacement"] = totalDisplacement

    # num_charts = length // 50

    for i in range(20):
        upper = (i + 1) * 50
        lower = i * 50
        chart = makeChartImage(
            trialData["SEQ"][lower:upper],
            trialData["total-displacement"][lower:upper],
            baseImage,
        )
        if random.randint(0, 100) >= 80:
            chart.save(f"testData/{chartType}Data/{file}-{i}.png")
        else:
            chart.save(f"trainData/{chartType}Data/{file}-{i}.png")
            
```
