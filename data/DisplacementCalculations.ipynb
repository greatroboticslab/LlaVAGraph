```python
delta = list(trialData["D-Value"])
deltaDValue = list(trialData["D-Value"])
deltaDisplacement = list(trialData["D-Value"])
totalDisplacement = list(trialData["D-Value"])
```

```python
deltaDValue[0] = 0
for i in range(1, len(delta)):
    deltaDValue[i] = delta[i] - delta[i-1]
deltaDValue[0] = 0
```

```python
deltaDValue
```

```python
deltaDisplacement[0] = 0
for i in range(1, len(delta)):
    deltaDisplacement[i] = (deltaDValue[i] + deltaDValue[i-1]) * 79.2
deltaDisplacement[0] = 0
```

```python
deltaDisplacement
```

```python
totalDisplacement[0] = 0
for i in range(1, len(delta)):
    totalDisplacement[i] = (deltaDisplacement[i] + totalDisplacement[i-1])
totalDisplacement[0] = 0
```

```python
totalDisplacement
```

```python
# trialData["Delta-D-Value"] = deltaDValue
trialData["Delta-Displacement"] = deltaDisplacement
trialData["Total-Displacement"] = totalDisplacement
```

```python
trialData
```
